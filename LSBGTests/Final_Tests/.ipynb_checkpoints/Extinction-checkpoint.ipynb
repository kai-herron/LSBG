```python
from sklearn.metrics import confusion_matrix

def rates(y_true, y_pred):
    CM = confusion_matrix(y_true, y_pred)

    TN = CM[0][0]
    FN = CM[1][0]
    TP = CM[1][1]
    FP = CM[0][1]
    
    # Initialize
    X_mat = np.zeros(4)
    
    # Populate
    X_mat[0] = TN
    X_mat[1] = FN
    X_mat[2] = TP
    X_mat[3] = FP
    
    return X_mat  
```

### Import feature matrix and labels

```python
X_feat_real = np.load('/data/des81.a/data/kherron/LSBG/trainingfiles/training_ext/X_feat_vis.npy')
y_lab_real = np.load('/data/des81.a/data/kherron/LSBG/trainingfiles/training_new/y_lab_new.npy')

X_y3 = np.load('/data/des81.a/data/kherron/LSBG/trainingfiles/training_ext/X_feat_y3.npy')
y_y3 = np.load('/data/des81.a/data/kherron/LSBG/trainingfiles/training/y_lab_y3.npy')

X_sims = np.load('/data/des81.a/data/kherron/LSBG/trainingfiles/training_ext/X_feat_sim.npy')
y_sims = np.ones(len(X_sims))

X_train_FINAL = np.concatenate((X_feat_real,X_y3,X_sims))
y_train_FINAL = np.concatenate((y_lab_real,y_y3,y_sims))
```

```python
#Split up LSBGs and aritfacts into train and test (70% train and 30% test)

index_real=(y_lab_real==1)
index_art=(y_lab_real==0)

#X_art_ind = np.random.choice(len(X_feat_real[index_art]),size=20000,replace=False)
#X_lsbg_ind = np.random.choice(len(X_feat_real[index_real]),size=20000,replace=False)

X_art = X_feat_real[index_art]#[X_art_ind]
X_lsbg = X_feat_real[index_real]#[X_lsbg_ind]

y_art = y_lab_real[index_art]#[X_art_ind]
y_lsbg = y_lab_real[index_real]#[X_lsbg_ind]


X_train_art, X_test_art, y_train_art, y_test_art = train_test_split(X_art, y_art,
                                                                        train_size = 0.70, random_state = 0)
X_train_lsbg, X_test_lsbg, y_train_lsbg, y_test_lsbg = train_test_split(X_lsbg, y_lsbg,
                                                                        train_size = 0.70, random_state = 0)

X_train_y3, X_test_y3, y_train_y3, y_test_y3 = train_test_split(X_y3, y_y3,
                                                                train_size = 0.70, random_state = 0)

X_train_sims, X_test_sims, y_train_sims, y_test_sims = train_test_split(X_sims, y_sims,
                                                                        train_size = 0.70, random_state = 0)


print('===========Train==Test')
print("ARTIFACTS:",len(X_train_art), len(X_test_art))
print("LSBGs    :",len(X_train_lsbg), len(X_test_lsbg))
print("Y3       :",len(X_train_y3), len(X_test_y3))
print("SIMs     :",len(X_train_sims), len(X_test_sims))
```

```python
#Create training arrays
X_train_norm = np.concatenate((X_train_art,X_train_lsbg,X_train_y3,X_train_sims))
y_train = np.concatenate((y_train_art,y_train_lsbg,y_train_y3,y_train_sims))

# Standardize the two sets
scaler = StandardScaler().fit(X_train_norm)
X_train = scaler.transform(X_train_norm)
```

```python
#Make the classifier
model_rf_fin = RandomForestClassifier(n_estimators=100)

model_rf_fin = model_rf_fin.fit(X_train,y_train)
```

```python
X_test_1 = np.concatenate((X_test_lsbg,X_test_art)) 
y_test = np.concatenate((y_test_lsbg,y_test_art))
```

```python
X_test = scaler.transform(X_test_1)

y_pred = model_rf_fin.predict(X_test)
```

```python
rate = rates(y_test,y_pred)
print("  TN      FN     TP     FP")
print(rate)
```

```python
TN = rate[0]
FN = rate[1]
TP = rate[2]
FP = rate[3]

# Sensitivity, hit rate, recall, or true positive rate
TPR_bin1_1 = TP/(TP+FN)
print("true positive rate:",TPR_bin1_1)
# Specificity or true negative rate
TNR_bin1_1 = TN/(TN+FP) 
print("true negative rate:",TNR_bin1_1)
# Precision or positive predictive value
PPV_bin1_1 = TP/(TP+FP)
print("positive predictive value:",PPV_bin1_1)
# Negative predictive value
NPV_bin1_1 = TN/(TN+FN)
print("negative predictive value:",NPV_bin1_1)
# Fall out or false positive rate
FPR_bin1_1 = FP/(FP+TN)
print("false positive rate:",FPR_bin1_1)
# False negative rate
FNR_bin1_1 = FN/(TP+FN)
print("false negative rate:",FNR_bin1_1)
# False discovery rate
FDR_bin1_1 = FP/(TP+FP)
print("false discovery rate:",FDR_bin1_1)

# Overall accuracy
ACC_bin1_1 = (TP+TN)/(TP+FP+FN+TN)
print("Overall accuracy:",ACC_bin1_1)
```

```python
from scipy.stats import ks_2samp
_ = plt.figure(figsize=(13,7))

X_evals = np.load('/data/des81.a/data/kherron/LSBG/trainingfiles/training_ext/X_feat_eval.npy')


bins = np.linspace(0,1)
dens,bins,bars = plt.hist(X_test_1[:,0],
                            bins=bins,histtype='step',
                            label='Visually Inspected Sample',
                            density=True,
                            stacked=True)

dens_e, bins_e,bars_e = plt.hist(X_evals[:,0],
                                   bins=bins,alpha=0.5,
                                   histtype='step',
                                   label='Evaluation Sample',
                                   density=True,
                                   stacked=True)
counts,bins = np.histogram(X_test_1[:,0],bins=bins)
counts_e, bins_e = np.histogram(X_evals[:,0], bins=bins)



plt.axhline(0,color='black')


bin_w = (max(bins) - min(bins)) / (len(bins) - 1)
plt.errorbar(bins[:-1]+(bin_w/2),dens, 
             yerr = (np.sqrt(counts)/(sum(counts)*np.diff(bins))), 
             ls='none')
plt.errorbar(bins[:-1]+(bin_w/2),dens_e, 
             yerr = (np.sqrt(counts_e)/(sum(counts_e)*np.diff(bins))), 
             ls='none')

plt.title('Ellipticity')
plt.legend(loc='upper right')
res = kstest(dens,dens_e)
plt.suptitle(t=('P-Value:'+str(res[1])),x=0.25,y=0.91)
plt.show()
```
